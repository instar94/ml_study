# 4. MySQL Query 실행

```python
mydb = mysql.connector.connect(
    host=host,
    user=user,
    password = password,
    database='zerobase'
)

mycursor = mydb.cursor()
### 테이블 생성 - 삭제
#mycursor.execute("create table sql_file (id int, name varchar(16))");
#mycursor.execute("drop table sql_file");

### SQL 파일을 실행 - 하나만 실행
#sql = open("query/test03.sql").read()
#mycursor.execute(sql)

### SQL 파일을 실행 - multi=True 옵션을 사용하여 여러 쿼리 실행
sql = open("query/test04.sql").read()
for result_iterator in mycursor.execute(sql, multi=True):
    if result_iterator.with_rows:
        print(result_iterator.fetchall())
    else:
        print(f"Query OK, {result_iterator.rowcount} rows affected.")
mydb.commit()
mydb.close()
```

# 5. CSV 데이터를 파이썬으로 INSERT

```python
import pandas as pd 
df = pd.read_csv("data/police_station.csv")
df.head()
```

```python
## 조회 쿼리 

mydb = mysql.connector.connect(
    host=host,
    user=user,
    password = password,
    database='zerobase'
)

mycursor = mydb.cursor(buffered=True)
sql = "INSERT INTO police_station VALUES (%s, %s)"

### for 문을 이용해 실행 
for i, r in df.iterrows():
    mycursor.execute(sql,tuple(r))
    print(tuple(r))
    mydb.commit()


mydb.close()
```

```python
## 조회 쿼리 

mydb = mysql.connector.connect(
    host=host,
    user=user,
    password = password,
    database='zerobase'
)

mycursor = mydb.cursor(buffered=True)
mycursor.execute("select * from police_station;")
result = mycursor.fetchall()
for data in result:
    print(data)
```
