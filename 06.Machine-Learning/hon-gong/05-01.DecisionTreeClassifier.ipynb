# 1. 결정 트리 
  - 문제 : 와인 데이터를 보고 와인 종류 맞추기 
  - 결정트리 : 설명이 `쉬운` 모델 
      - 스무 고개와 같은 모델 -> ㅇㅇ인가 Yes/No 로구분 

# 2. 로지스틱 회귀로 분류의 한계
## 1) 데이터 불러오기 
- 이진 분류 문제 
    - 0 : 레드와인 
    - 1 : 화이트 와인 

```python
import pandas as pd

wine = pd.read_csv("https://bit.ly/wine_csv_data")
wine.head()
```

  - 데이터 정보 확인 

```python
wine.info()
```

- 통계값 확인 
    - 각 컬럼마다 스케일이 다름 -> 표준화 후 사용이 필요 

```python
wine.describe()
```

- 데이터 와 타겟 나누기

```python
data = wine[["alcohol","sugar","pH"]]
target = wine["class"]
```

## 2) 훈련 세트와 테스트 세트로 분리
- `test_size` : 샘플을 몇 퍼센트 테스트 세트로 분리할지 지정 

```python

from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(data, target, random_state=42,test_size=0.2)

print("train_input.shape:", train_input.shape)
print("test_input.shape:", test_input.shape)
```

## 3) 데이터 표준화 

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

## 4) 로지스틱 회귀 모델을 이용한 추측 
- 결과가 좋지 않음 
  - 훈련 (0.78) +  검증 (0.77) 모두 낮음  ->  과소적합

```python
from sklearn.linear_model import LogisticRegression
lr =LogisticRegression()

### 모델 학습
lr.fit(train_scaled, train_target)

### 모델 평가
print("모델 평가 (훈련) : " ,lr.score(train_scaled, train_target))
print("모델 평가 (검증) : " ,lr.score(test_scaled, test_target))
```

# 3. 결정 트리 (Decision Tree) 로 분류하기 

## 1) 결정 트리를 통한 분류 학습 

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=42)
### 모델 학습
dt.fit(train_scaled, train_target)

### 모델 평가
print("모델 평가 (훈련) : " ,dt.score(train_scaled, train_target))
print("모델 평가 (검증) : " ,dt.score(test_scaled, test_target))    
```

- 훈련세트에 대한 점수 상승 
- 과대 적합 문제 발생 (훈련(99) > 검증(86))

## 2) 결정 트리 시각화 
- `plot_tree()` 함수를 이용해서 사용 가능 
- `루트 노드`  : 가장 위에 있는 노드
- `리프 노드` : 가장 아래에 있는 노드

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(10,7))
plot_tree(dt)
plt.show()
```

- `max_depth`  : 트리의 깊이 제한 
- `filled ` : 노드 색상을 채우기 여부

```python
plt.figure(figsize=(10,7))
plot_tree(
    dt,
    max_depth=1,
    filled=True,
    feature_names=(["alcohol", "sugar", "pH"])
)
plt.show()
```

## 3) 불순도 (gini)
### gini 불순도 
- 결정 트리가 최적의 질문을 찾기위한 기준
    -  사이킷런은 지니불순도와 엔트로피 불순도를 이용
- 지니 불순도 식  : $\text(지니불순도)=1-(\text(음성클래스 비율)^2 + \text(양성클래스 비율)^2)$
- 불순도 0.5 = 가장 불순함 
  
  ### 정보 이득
  - 결정 트리 모델은 부모 노드와 자식 노드의 불순도 차이 가 가능한 크도록 트리를 만듦
  - 자식노드와 부모 노드 사이 불순도 차이 
  - 결정 트리 모델은 정보이득을 최대화 하도록 함 

## 4) 가지 치기
- 성장을 제한 하는 방법 
-  가지치기를 하지 않으면 훈련 세트에만 잘 맞게됨 -> 일반화가 잘 되지 않음 

### 가치치기 1 : 트리 최대 깊이 설정 
`max_depth` : 가지 깊이 설정 
  - 훈련 세트는 점수는 낮아졌지만 테스트 성능은 거의 유지

```python
dt = DecisionTreeClassifier(max_depth=3, random_state=42)
### 모델 학습
dt.fit(train_scaled, train_target)  
### 모델 평가
print("모델 평가 (훈련) : " ,dt.score(train_scaled, train_target))
print("모델 평가 (검증) : " ,dt.score(test_scaled, test_target))
```

- 시각화 
  - 깊이 1  : 당도(sugar)
  - 깊이 2 : 알콜 도수 

```python
plt.figure(figsize=(10,7))
plot_tree(dt, filled=True, feature_names=(["alcohol", "sugar", "pH"]))
plt.show()
```

## 5) 전처리 불필요 
- 결정트리는 표준화 전처리가 필요하지 않음 
  - 샘플을 어떤 클래스 비율로 나누는지 계산 할때 특성 값의 스케일이 계산에 영향을 미치지 않음

```python
dt = DecisionTreeClassifier(max_depth=3, random_state=42)
### 모델 학습
dt.fit(train_input, train_target)
### 모델 평가
print("모델 평가 (훈련) : " ,dt.score(train_input, train_target))
print("모델 평가 (검증) : " ,dt.score(test_input, test_target))
```

- 시각화 

```python
plt.figure(figsize=(10,7))
plot_tree(dt, filled=True, feature_names=(["alcohol", "sugar", "pH"]))
plt.show()
```

## 6) 특성 중요도 
- 결정 트리에 사용된 특성이 불순도를 감소하는데 기여한 정도
- `dt.feature_importances_` : 특성 중요도 출력 

```python
print("특성 중요도 : ", dt.feature_importances_)
```
