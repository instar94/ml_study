# 1. 아이리스 데이터 관찰하기
- `sklearn`의 `datasets` 모듈에서 `load_iris` 함수를 사용해 iris 데이터를 로드

```python
from sklearn.datasets import load_iris
iris = load_iris()
print(type(iris))
```

```python
iris.keys()
```

  - 데이터의 특성을 보여줌

- 머신러닝을 통해 맞춰야 하는 것들  (=정답 데이터)

### feature 
- 머신러닝에 이용 할 특성들 

```python
print(iris["DESCR"])
```

### target
- 정답 데이터 
- `target` 키는 각 데이터가 어떤 정답을 가지고 있는지 인덱스로 표현 
- `target_names`는 각 인덱스(=라벨)이 어떤걸 의미하는지 표현

```python
print(iris["target"] )
```

```python
print(iris["target_names"] )
```

### data
- 특성 데이터 

```python
print(iris["data"])
```

# 2. 판다스를 이용해 DataFrame으로 변환

```python
import pandas as pd
```

```python
iris_feature = iris["feature_names"]
iris_data = iris["data"]

iris_pd = pd.DataFrame(iris_data,columns=iris_feature)
iris_pd
```

## 품종 데이터를 추가함 

```python
iris_pd["species"] = iris["target"]
iris_pd.head()
```

## 시각화 
- 박스 플롯으로 만드어 각 종별 특징 확인 

```python
import matplotlib.pyplot as plt
import seaborn as sns  
```

- sepal length는 겹치는 부분이 존재함 
- sepal length만으로는 구분이 어려움 
- 2번의 아웃라이어가....

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="sepal length (cm)", y="species", data=iris_pd, orient="h", hue="species", palette="Set1" )
```

### 꽃잎 특성 
- 0의 경우 확실하게 구분 할 수 있음 

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="petal length (cm)", y="species", data=iris_pd, orient="h", hue="species", palette="Set1" )
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="petal width (cm)", y="species", data=iris_pd, orient="h", hue="species", palette="Set1" )
```

## pairplot으로 시각화 

```python
sns.pairplot(iris_pd, hue="species", palette="Set1")
```

- 이중에서 petal length와 petal width의 관계 집중 

```python
sns.pairplot(iris_pd,
             vars=["petal length (cm)", "petal width (cm)"],
             hue="species",
             palette="Set1",
             height=4)
```

### Decision Tree 
- 0번 데이터의 경우, 확실하게 구분 가능 
- 1과 2의 경우, 겹치는 부분이 존재
  - 몇 개의 데이터는 틀릴 수 있지만 구분은 가능 
  - 꽃잎 길이가 2.5 이상 + width가 1.6보다 작음 -> 1
  - 꽃잎 길이가 2.5 이상 + width가 1.6보다 크면 -> 2

--> 이런 내용의 질문으로 순차적으로 표현이 가능한가 ? 
  --> Decision Tree 알고리즘
