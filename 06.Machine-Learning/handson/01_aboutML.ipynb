# 모델 기반 학습 예시

## 1. 가설 
> 돈이 사람을 행복하게 만든다. 

## 2. 패키지 임포트 

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

## 3. 데이터 불러오기 

- 데이터 경로 선언 후

```python
### 깃허브 데이터 경로 선언
data_root = "https://github.com/ageron/data/raw/main"
lifesat = pd.read_csv(data_root + "/lifesat/lifesat.csv")
```

### 데이터 확인 
- GDP per capita (USD) : 1인당 GDP
- Life satisfaction : 행복도 

```python
lifesat.head()
```

## 4.  시각화를 통한 데이터 확인
  - 산점도로 데이터의 전체적인 모습을 확인 
    - `kind="scatter"` : 산점도 그리기 

- 데이터 확인 : 선형이 형성될 것으로 보임 

```python
lifesat.plot(kind="scatter", ### 산점도
            x="GDP per capita (USD)", ### x축
            y="Life satisfaction") ### y축

###  [x축 시작점, x축 끝점, y축 시작점, y축 끝점]
plt.axis([23_500, 62_500, 4, 9 ])

### 산점도 그리기 
plt.show()
```

## 5. 선형 회귀 모델 생성 

### 학습 데이터 생성 
- X : 데이터 (알아 맞추는 데이터)
- Y : 레이블 (알아 맞춰야 하는 데이터)

```python
X = lifesat[["GDP per capita (USD)"]].values
y = lifesat["Life satisfaction"].values
```

### 선형 모델 생성 및 훈련 

```python
### 선형 회귀 모델 선택 
model = LinearRegression()

### 모델 훈련
model.fit(X, y)
```

```python
### 산점도 다시 그리기
lifesat.plot(kind="scatter", 
            x="GDP per capita (USD)", 
            y="Life satisfaction")

### 산점도에 선형 회귀선 그리기
plt.plot(X, model.predict(X), color="red", linewidth=2, label="Linear regression")


### 축 범위 설정
plt.axis([23_500, 62_500, 4, 9 ])

plt.show()
```

### 모델을 활용한 데이터 예측 

```python
X_new = [[37_655.2]]
print("예측 행복도 : " ,model.predict(X_new) )
```
